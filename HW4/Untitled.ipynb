```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
```

```python
t, s = sympy.symbols('t, s')
```

```python
J=0.1
b=1
K=10
```

```python
F=((J*s+b)*(0.5*s+1))/((0.5*s+1)*(J*s+b)+10*K)
```

```python
f=sympy.inverse_laplace_transform(F, s, t)
```

```python
T=np.arange(0.00000001,2,0.01)
Y=[]
for i in T:
    Y.append(f.subs(t,i))
```

```python
plt.plot(T,Y)
```
