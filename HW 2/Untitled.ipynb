```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=[]
Y=[]

x=-np.arange(0.864,5,0.01)[::-1]
y=x*np.log(0.16)/np.pi
print(f"{x[0]},{x[-1]}")
X+=x.tolist()
Y+=y.tolist()

y=np.arange(-0.504,0.504,0.01)[::-1]
x=-np.sqrt(1-y**2)
print(f"{x[0]},{x[-1]}")
X+=x.tolist()
Y+=y.tolist()

x=-np.arange(0.864,5,0.01)
y=-x*np.log(0.16)/np.pi
print(f"{x[0]},{x[-1]}")
X+=x.tolist()
Y+=y.tolist()
```

```python

plt.fill_betweenx(Y,X,-5)
plt.ylim(-5,5)
plt.xlim(-5,5)
plt.xlabel("Re(s)")
plt.ylabel("Im(s)")
plt.savefig("Problem2")
```
