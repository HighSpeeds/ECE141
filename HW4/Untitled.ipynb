```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
```

```python
t, s = sympy.symbols('t, s')
```

```python
J=0.1
b=1
K=10
```

```python
F=((J*s+b)*(0.5*s+1))/((0.5*s+1)*(J*s+b)+10*K)
```

```python
f=sympy.inverse_laplace_transform(F, s, t)
```

```python
T=np.arange(0.00000001,2,0.01)
Y=[]
for i in T:
    Y.append(f.subs(t,i))
```

```python
plt.plot(T,Y)
```

```python
a=0.5*0.1
b=0.6
c=101

np.sqrt(4*a*c-b**2)
```

```python
roots=np.array([(-b+np.sqrt(4*a*c-b**2)*1j)/(2*a),
       (-b-np.sqrt(4*a*c-b**2)*1j)/(2*a)])
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,5))

plt.sca(axs[0])
plt.plot(np.real(roots),np.imag(roots),"x")
plt.xlim(-10,0)
plt.xlabel("real(s)")
plt.ylabel("imag(s)")
plt.title("roots")

plt.sca(axs[1])
plt.plot(T,Y)
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("Step Response")
plt.savefig("fig1")
```
